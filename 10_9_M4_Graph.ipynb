**Graph**\
A graph is a data structure that consists of a set of nodes (or vertices) connected by edges. Graphs are widely used to model relationships in various domains, such as social networks, transportation systems, and computer networks. Here‚Äôs an overview of the key components, types, and applications of graphs.

Key Components
1. **Vertices (Nodes)**: the fundamental units of a graph. Each node can represent an entity, such as a person in a social network or a location in a transportation map.
2. **Edges**: the connections between nodes. An edge can be directed (indicating a one-way relationship) or undirected (indicating a two-way relationship). Edges may also have weights, representing costs, distances, or other metrics.


**Types of Graphs**
1. **Undirected Graph:** a graph where the edges have no direction. The connection between two vertices is bidirectional.
2. **Directed Graph (Digraph):** a graph where edges have a direction, indicated by arrows. The connection from vertex A to vertex B does not imply a connection from B to A.
3. **Weighted Graph:** a graph in which edges have associated weights, representing costs, lengths, or capacities.
4. **Unweighted Graph:** a graph where edges do not have weights. The connections are simply present or absent.
5. **Simple Graph:** a graph with no loops (edges connecting a vertex to itself) and no multiple edges between two vertices.
6. **Cyclic Graph:** a graph that contains at least one cycle (a path where the start and end nodes are the same).
7. **Acyclic Graph:** a graph that does not contain any cycles.

**Applications of Graphs**
1. **Social Networks:** representing relationships between users (friends, followers) as nodes and edges.
2. **Transportation Networks:** modeling routes and connections between locations (cities, bus stops).
3. **Computer Networks:** representing connections between devices, such as routers and switches.
4. **Recommendation Systems:** analyzing user-item interactions to suggest products or content.

**Graph Representation**
Graphs can be represented in various ways:\
1. **Adjacency Matrix:** a 2D array where the cell at row ùëñ and column ùëó indicates the presence (and possibly the weight) of an edge from vertex
ùëñ to vertex ùëó.
2. **Adjacency List:** a list where each vertex has a corresponding list of adjacent vertices. This representation is often more space-efficient for sparse graphs.



[**NetworkX**](https://networkx.org/documentation/stable/reference/functions.html) is a powerful Python library designed for the creation, manipulation, and study of complex networks (graphs). It provides tools for the representation of both undirected and directed graphs, along with various algorithms for analysis and visualization. Below is an overview of its features, installation, and some common use cases.

Key Features of NetworkX\
**Graph Types:** supports undirected graphs, directed graphs, multigraphs (with multiple edges between nodes), and weighted graphs.

**Graph Algorithms:** implements numerous graph algorithms, including shortest path algorithms, clustering, centrality measures, connectivity analysis, and community detection.

**Visualization:** basic visualization capabilities integrated with Matplotlib, allowing you to draw graphs and customize their appearance.

**Input/Output:**can read and write graphs in various formats (e.g., GraphML, GML, Pajek).

**Dynamic Graphs: **supports dynamic graph structures, enabling the addition and removal of nodes and edges easily.

```python
import networkx as nx                # Import the networkx library for graph operations
import matplotlib.pyplot as plt       # Import matplotlib for graph visualization

# Create an undirected graph
G = nx.Graph()                       # Initialize an empty undirected graph

# Add nodes
G.add_node(1)                        # Add node 1 to the graph
G.add_node(2)                        # Add node 2 to the graph
G.add_node(3)                        # Add node 3 to the graph

# Add edges
G.add_edges_from([(1, 2), (1, 3), (2, 3)])  # Add edges between nodes

# Draw the graph
nx.draw(G, with_labels=True)         # Draw the graph with node labels
plt.title("Simple Undirected Graph")  # Set the title for the plot
plt.show()                            # Display the plot
```

```python
# Calculate and print basic properties of directed graph
print("Number of nodes:", G.number_of_nodes())  # Print the total number of nodes in the graph
print("Number of edges:", G.number_of_edges())  # Print the total number of edges in the graph
print("Degree of node 2:", G.degree[2])         # Print the degree (number of connections) of node 2
print("Is the graph connected?", nx.is_connected(G))  # Check if the graph is connected and print the result
```

```python
import networkx as nx                # Import the networkx library for graph operations
import matplotlib.pyplot as plt       # Import matplotlib for graph visualization
# Create a sample graph
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)])

# Define positions for the nodes
pos = nx.spring_layout(G)

# Draw the nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700)

# Draw all edges with default style
nx.draw_networkx_edges(G, pos)

# Highlight specific edges (e.g., between nodes 1 and 2, and 3 and 4)
highlighted_edges = [(1, 2), (3, 4)]
nx.draw_networkx_edges(G, pos, edgelist=highlighted_edges, edge_color='red', width=2)

# Set title and show the plot
plt.title("Graph with Highlighted Edges")
plt.show()
```

```python
import networkx as nx                # Import the networkx library for graph operations
import matplotlib.pyplot as plt       # Import matplotlib for graph visualization

# Create an undirected graph
G = nx.Graph()                       # Initialize an empty undirected graph
G.add_nodes_from([1, 2, 3,4,5])
# Draw the graph
nx.draw(G, with_labels=True)         # Draw the graph with node labels
plt.title("Simple Unconnected Graph")  # Set the title for the plot
plt.show()
```

```python
# Calculate and print basic properties of directed graph
print("Number of nodes:", G.number_of_nodes())  # Print the total number of nodes in the graph
print("Number of edges:", G.number_of_edges())  # Print the total number of edges in the graph
print("Degree of node 2:", G.degree[2])         # Print the degree (number of connections) of node 2
print("Is the graph connected?", nx.is_connected(G))  # Check if the graph is connected and print the result
```

```python
# Create a directed graph
DG = nx.DiGraph()                    # Initialize an empty directed graph

# Add nodes and edges
DG.add_edges_from([(1, 2), (2, 3), (3, 1),(3,2)])  # Add directed edges between nodes

# Draw the directed graph
nx.draw(DG, with_labels=True, arrowstyle='-|>', arrowsize=20)  # Draw the directed graph with arrows
plt.title("Simple Directed Graph")    # Set the title for the plot
plt.show()                            # Display the plot
```

```python
# Calculate and print basic properties of directed graph
print("Number of nodes:", DG.number_of_nodes())  # Print the total number of nodes in the graph
print("Number of edges:", DG.number_of_edges())  # Print the total number of edges in the graph
print("Degree of node 2:", DG.degree[2])         # Print the degree (number of connections) of node 2
```

```python
# Create a weighted graph
WG = nx.Graph()                         # Initialize an empty undirected graph for weighted edges

# Add edges with weights
WG.add_weighted_edges_from([(1, 2, 4), (1, 3, 2), (2, 3, 1)])  # Add edges with weights

# Draw the weighted graph
pos = nx.spring_layout(WG)              # Generate positions for nodes in a visually appealing way
nx.draw(WG, pos, with_labels=True)      # Draw the graph with node labels
labels = nx.get_edge_attributes(WG, 'weight')  # Get the weights of the edges
nx.draw_networkx_edge_labels(WG, pos, edge_labels=labels)  # Draw the edge labels (weights)
plt.title("Weighted Graph")             # Set the title for the plot
plt.show()                              # Display the plot
```

```python
# Create a simple graph
G = nx.Graph()                         # Initialize an empty undirected graph
G.add_weighted_edges_from([(1, 2, 4), (1, 3, 2), (2, 3, 1), (3, 4, 5)])  # Add weighted edges

# Find the shortest path from node 1 to 4
shortest_path = nx.shortest_path(G, source=1, target=4, weight='weight')  # Calculate the shortest path
print("Shortest path from 1 to 4:", shortest_path)  # Print the shortest path

# Draw the graph with the shortest path highlighted
pos = nx.spring_layout(G)              # Generate positions for nodes
nx.draw(G, pos, with_labels=True)      # Draw the graph with node labels
nx.draw_networkx_edges(G, pos, edgelist=[(1, 2), (2, 3), (3, 4)], edge_color='r', width=2)  # Highlight the shortest path
plt.title("Graph with Shortest Path Highlighted")  # Set the title for the plot
plt.show()                              # Display the plot
```
