The binarytree moduleLinks to an external site.in Python provides a convenient way to create and manipulate binary trees. It supports various functionalities such as constructing trees from lists, accessing node properties, and performing traversals. 

Use  the binarytree module to build the tree below:



```python
# install it: 
#!pip install binarytree
```

```python
import binarytree
from binarytree import Node 

root = Node(3)
root.left = Node(1)
root.right = Node(6)
# left side
root.left.left = Node(2)
root.left.right = Node(11)
#right side
root.right.left=Node(9)
root.right.right=Node(5)
# print the tree
print(root)
```
